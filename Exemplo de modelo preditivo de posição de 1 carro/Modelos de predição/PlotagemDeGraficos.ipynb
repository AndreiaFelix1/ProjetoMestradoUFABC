```python
import pandas as pd

caminho_arquivo_excel = "/content/drive/MyDrive/ProjetoVeins/Plotagem5Veiculos/Resultadosgerais5veiculos.xlsx"

df = pd.read_excel(caminho_arquivo_excel)

print(df.head())
```

```python
df
```

```python
# Simplificar numero de posição
lista = []
for i in range(0,len(df)):
  v = str(df['POSITION'][i]).split(',')
  item = str(v[0]).replace('(','')
  lista.append(item)
print(lista)

df['POSITION FRONT'] = lista
```

```python
df
```

```python
# Inicializando uma lista para armazenar as séries de contagem
series_contagem = []

# Inicializando um dicionário para manter o contador para cada valor de 'id'
contador_por_id = {}

# Loop sobre as linhas do DataFrame
for indice, linha in df.iterrows():
    id_atual = linha['ID']

    # Se o id atual não existir no dicionário, inicializa o contador para 1
    if id_atual not in contador_por_id:
        contador_por_id[id_atual] = 1
    else:
        # Caso contrário, incrementa o contador
        contador_por_id[id_atual] += 1

    # Adiciona o valor do contador à lista
    series_contagem.append(contador_por_id[id_atual])

df['TEMPO'] = series_contagem
print(df)
```

```python
print(df)
```

```python
import matplotlib.pyplot as plt

ids_unicos = df['ID'].unique()

for id_unico in ids_unicos:
    filtro = df[df['ID'] == id_unico]

    plt.scatter(filtro['TEMPO'], filtro['SPEED'], label=f'ID {id_unico}')

    plt.xlabel('Tempo')
    plt.ylabel('Velocidade')

    plt.title(f'Gráfico de velocidade em função do tempo para o veículo ID {id_unico}')

    plt.legend()
    plt.show()
```
