```python
import matplotlib.pyplot as plt

# Dados organizados conforme fornecido
modelos = ['ARIMA', 'KNN REGRESSOR', 'SVR LINEAR', 'SVR RBF', 'SVR POLINOMIAL', 'REGRESSÃO LINEAR']

mae = [14.09, 8.68, 5.19, 13.41, 135.32, 2.01]
mse = [3.06, 18.90, 71.95, 481.79, 0, 6.90]
rmse = [17.50, 137.48, 8.48, 21.95, 159.05, 2.63]

# MSE SVR POLINOMIAL 25298.23

# Criar figura e eixos
fig, ax = plt.subplots(figsize=(10, 6))

# Plotar barras para MAE, MSE e RMSE
bar_width = 0.25
index = range(len(modelos))

bar1 = ax.bar(index, mae, bar_width, label='MAE')
bar2 = ax.bar([i + bar_width for i in index], mse, bar_width, label='MSE')
bar3 = ax.bar([i + 2 * bar_width for i in index], rmse, bar_width, label='RMSE')

# Adicionar legendas, título e rótulos nos eixos
ax.set_xlabel('Modelos')
ax.set_ylabel('Erro')
ax.set_title('Comparação de MAE, MSE e RMSE entre Modelos')
ax.set_xticks([i + bar_width for i in index])
ax.set_xticklabels(modelos, rotation=45)
ax.legend()

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
